### 1. Генерация сигналов

#### Определение принадлежности к пространствам $L_1(R), L_2(R), S$ и описание свойств сигнала и его преобразования Фурье:

В указанном сигнале экспонента вносит ключевой вклад, заставляя функцию быстро убывать на бесконечности.
Поэтому:
1) $L_1$ (Несобственный интеграл от модуля функции конечен): <br>
Сигнал непрерывен на всей оси (нет деления на ноль в знаменателе, он $\ge$ 4). Экспоненциальное затухание делает площадь под графиком конечной. Поэтому несобственный интеграл от модуля функции сходится. <br>
**Вывод: $x_2(t) \in L_1$**
2) $L_2$ (Несобственный интеграл от квадрата модуля конечен): <br>
Если возвести функцию в квадрат, она будет затухать еще быстрее. <br>
**Вывод: $x_2(t) \in L_2$**
3) $S$ (Бесконечно дифференцируемые быстроубывающие функции): <br>
Наличие модуля в функции приводит к излому в точках $t = -1, t = 2$, поэтому функция хоть и непрерывна, но не имеет даже первой производной в этих точках. <br>
**Вывод: $x_2(t) \notin S$**

Свойства сигнала и его Преобразования Фурье:
1) По лемме Римана — Лебега спектр сигнала является непрерывной и ограниченной функцией, и при $\lambda \to \pm \infty$, $\hat{x}(\lambda) \to 0$, так как $x_2(t) \in L_1$
2) По равенству Планшереля - Парсеваля энергия сигнала в частотной области равна энергии во временной, так как $x_2(t) \in L_2$ и $\hat{x_2}(t) \in L_2$


```python
# необходимые фреймворки
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ВХОДНЫЕ ДАННЫЕ
# задание параметров
a2 = -1.0
b2 = 4.0
t2 = -1.0
c2 = 0.2
t3 = 2.0
d2 = 11.0

tmin, tmax = -10, 10
N = 1000 # количество дискретных шагов
t = np.linspace(tmin, tmax, N)

# задание функции
x_func = lambda t: (a2 / (b2 + np.abs(t - t2))) * np.exp(-c2 * np.abs(t - t3)) * np.sinc(d2 * t)
```

```python
def plot_signal(func, tmin=-10, tmax=10, N=1000, figsize=(10, 4), title="График сгенерированного сигнала во времени"):
  t = np.linspace(tmin, tmax, N)
  x=func(t)
  plt.figure(figsize=figsize)
  plt.plot(t, x, color='blue', linewidth=1.5, label='Сигнал $x_2(t)$')
  plt.title(title)
  plt.xlabel('Время $t$, с')
  plt.ylabel('Амплитуда $x(t)$')
  plt.axhline(0, color='black', linewidth=0.5)
  plt.grid(True, linestyle='--', alpha=0.7)
  plt.legend()
  plt.tight_layout()
  plt.show()
```

```python
# визуализация графика с параметрами по дефолту (пример 1)
plot_signal(x_func, title="График сгенерированного сигнала $x_2(t)$ во времени, #1")
```

```python
# визуализация графика с изменёнными параметрами, чтобы рассмотреть центральные вершины поближе (пример 2)
plot_signal(x_func, tmin=-2.5, tmax=2.5, N=500, title="График сгенерированного сигнала $x_2(t)$ во времени, #2")
```
